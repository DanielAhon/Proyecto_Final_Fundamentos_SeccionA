```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Sobre Groupby
#El proceso de groupby se puede dividir en tres pasos principales:

# 1. Dividir (Split):
# El DataFrame se divide en grupos basados en los valores únicos de una o más columnas especificadas.

# 2.Aplicar (Apply):
# Se aplica una función a cada grupo de forma independiente. Esta función puede ser de agregación (como sum, mean, median), transformación (como normalización), o filtro (como eliminar grupos basados en un criterio).

# 3. Combinar (Combine):
# Los resultados de la función aplicada a cada grupo se combinan de nuevo en una estructura de datos.

# Crear el DataFrame
scores = pd.DataFrame({
    'name':['Adam', 'Bob', 'Dave', 'Fred'],
    'age': [15, 16, 16, 15],
    'test1': [95, 81, 89, None],
    'test2': [80, 82, 84, 88],
    'teacher': ['Ashby', 'Ashby', 'Jones', 'Jones']
})

# Agrupar por maestro y calcular la mediana
print("Separando por maestro: ")
grupo = scores.groupby('teacher').median(numeric_only=True)
display(grupo)

# Esto incluyó la columna age, para ignorar que podemos separar solo las columnas de prueba
print("Separar las columnas de prueba(test), pues incluyo edad")
agrupar_tests = scores[['test1', 'test2', 'teacher']].groupby('teacher').median(numeric_only=True)
display(agrupar_tests)

# Agrupar por maestro y edad, y calcular la mediana
print("Maestro y edad, mediana:")
aguarar_por_maestro_edad = scores[['teacher', 'age']].groupby('teacher').median(numeric_only=True)
display(aguarar_por_maestro_edad)

# Obtener los puntajes mínimos y máximos de las pruebas por maestro, .agg() llama a una lista de funciones
print("Puntajes minimo y maximo de cada maestro")
grupo_agg = scores.groupby('teacher').agg({'test1': ['min', 'max'], 'test2': ['min', 'max']})
display(grupo_agg)
```

```python
#TABLAS PIVOT , similar a una tabla dinamica en excel


# Cargar el archivo CSV
iris = pd.read_csv(lectura + "iris.csv")

# Media de las medidas de los sépalos y pétalos por especie
pivot_media = iris.pivot_table(index='Species', aggfunc='mean')
print("Media de las medidas de los sépalos y pétalos por especie:")
display(pivot_media)

# Medidas mínimas y máximas de los sépalos y pétalos por especie
pivot_min_max = iris.pivot_table(index='Species', values=["Sepal.Length","Sepal.Width"], aggfunc=[min, max]) #values: solo de aquellas columnas
print("\nMedidas mínimas y máximas de los sépalos y pétalos por especie:")
display(pivot_min_max)

# Múltiples estadísticas por especie,
pivot_multiple = iris.pivot_table(index='Species', aggfunc={
    'Sepal.Length': ['mean', 'std', 'sum'],
    'Sepal.Width': ['mean', 'std', 'sum'],
    'Petal.Length': ['mean', 'std', 'sum'],
    'Petal.Width': ['mean', 'std', 'sum']
})
print("\nMúltiples estadísticas por especie:")
display(pivot_multiple)

# Agregar filas de resumen
pivot_con_margenes = iris.pivot_table(index='Species', aggfunc='mean', margins=True ) #margins: True -> Agrega una fila adicional, resultado de la opeacion(agregación) realizada
print("\nMedia de las medidas de los sépalos y pétalos por especie con filas de resumen:")
display(pivot_con_margenes)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo 1: Concatenación básica
x = np.array([4, 5, 6])
y = np.array([7, 8, 9])
print("Concatenación básica:")
print(np.concatenate([x, y]))

# Ejemplo 2: Concatenación de dos o más matrices
z = [99, 99, 99]
print("Concatenación de tres arrays:")
print(np.concatenate([x, y, z]))

# Ejemplo 3: Concatenando una matriz dos veces
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("Concatenación de una matriz consigo misma:")
print(np.concatenate([grid, grid]))

# Ejemplo 4: Usando vstack para matrices de distintas dimensiones
x = np.array([0, 1, 2])
grid = np.array([[3, 4, 5],
                 [6, 5, 4]])
print("Concatenación vertical (vstack):")
print(np.vstack([x, grid]))

# Ejemplo 5: Usando hstack para matrices de distintas dimensiones
y = np.array([[23],
              [23]])
print("Concatenación horizontal (hstack):")
print(np.hstack([grid, y]))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo inicial de suma de arrays
a = np.array([0, 1, 2])
b = np.array([2, 6, 1])
print("Suma de arrays:")
print(a + b)

# Broadcasting con escalar
print("Broadcasting con escalar:")
print(a + 5)

# Broadcasting con matrices de diferentes dimensiones
A = np.arange(16).reshape(4, 4)
b = np.arange(4)
print("Matriz A:")
print(A)
print("\nVector b:")
print(b)
print("\nSuma A + b (broadcasting):")
print(A + b)

# Ejemplos adicionales de broadcasting
# Ejemplo 1: Broadcasting con matrices de diferentes formas
M = np.ones((3, 3))
v = np.arange(3)
print("Matriz M:")
print(M)
print("\nVector v:")
print(v)
print("\nSuma M + v (broadcasting):")
print(M + v)

# Ejemplo 2: Broadcasting en múltiples dimensiones
a = np.arange(3).reshape((3, 1))
b = np.arange(3)
print("Matriz a:")
print(a)
print("\nVector b:")
print(b)
print("\nSuma a + b (broadcasting):")
print(a + b)

# Ejemplo 3: Broadcasting con matrices de diferentes dimensiones
X = np.random.random((10, 3))
w = np.random.random((3,))
print("Forma de X:", X.shape)
print("Forma de w:", w.shape)
print("\nForma del resultado de X * w:", (X * w).shape)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo de broadcasting con J
J = np.arange(0, 40, 10)
print("J original:")
print(J)
print("Forma de J:", J.shape)

J = J[:, np.newaxis]  # agregamos un nuevo eje ->  matriz 2D
print("\nJ después de agregar un nuevo eje:")
print(J)
print("Nueva forma de J:", J.shape)

print("\nJ + 3 (broadcasting):")
print(J + 3)

# Comparaciones y lógica booleana
a1 = np.array([1, 2, 3, 4])
b1 = np.array([4, 3, 2, 1])
print("\nComparación a1 < b1:")
print(a1 < b1)

print("\nnp.all(a1 < b1):", np.all(a1 < b1))
print("np.any(a1 < b1):", np.any(a1 < b1))

if np.all(a1 < b1):
    print("Todos los elementos en a1 son menores que los elementos en b1")
elif np.any(a1 < b1):
    print("Algunos elementos en a1 son menores que los elementos de b1")
else:
    print("Todos los elementos en b1 son menores que los elementos de a1")

# Conversión de booleanos a enteros
print("\nConversión de booleanos a enteros:")
print("Suma de booleanos:", np.sum(a1 < b1))
print("Media de booleanos:", np.mean(a1 < b1))

# Operaciones con matrices booleanas
print("\nOperaciones con matrices booleanas:")
c = np.array([1, 0, 1, 0, 1, 1, 0, 1])
print("Matriz original:", c)
print("Invertir booleanos:", ~c)
print("AND lógico:", np.logical_and(c, c))
print("OR lógico:", np.logical_or(c, ~c))

# Función de pulso usando operaciones booleanas
def pulso(t, posicion, altura, ancho):
    return altura * (t >= posicion) * (t <= (posicion + ancho))

t = np.linspace(0, 10, 100)
print("\nEjemplo de función de pulso:")
print(pulso(t, 4, 2, 1))

# Matrices booleanas como máscaras
k = np.array([1, 3, -1, 5, 7, -1])
mask = (k < 0)
print("\nMáscara booleana:")
print(mask)

# Ejemplo bidimensional
rng = np.random.RandomState(0)
z1 = rng.randint(10, size=(3, 4))
print("\nMatriz aleatoria z1:")
print(z1)

mask = (z1 > 5)
print("\nMáscara para z1 > 5:")
print(mask)

print("\nValores de z1 donde z1 > 5:")
print(z1[mask])

print("\nSuma de la máscara (cuenta de valores > 5):")
print(np.sum(mask))

print("\nMedia de la máscara (fracción de valores > 5):")
print(np.mean(mask))
```

```python
#MATPLOTIB
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {
    'A': np.random.randint(1, 100, 10),
    'B': np.random.randint(1, 100, 10),
    'C': np.random.randint(1, 100, 10),
    'D': np.random.randint(1, 100, 10)
}
df = pd.DataFrame(data)

# Descripción del DataFrame
print("Descripción del DataFrame:")
print(df.describe())

# Añadir una nueva columna E
df['E'] = df['A'] + df['B']

# Filtrar filas donde la columna A es mayor que 50
filtered_df = df[df['A'] > 50]

# Ordenar el DataFrame por la columna C
sorted_df = df.sort_values(by='C')

print("\nDataFrame con nueva columna E:")
print(df)

print("\nFiltrado (A > 50):")
print(filtered_df)

print("\nOrdenado por C:")
print(sorted_df)

# Visualización de datos

# Histograma de la columna A
plt.figure(figsize=(10, 5))
plt.hist(df['A'], bins=10, alpha=0.7, color='blue')
plt.title('Histograma de la columna A')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Gráfico de dispersión entre A y B
plt.figure(figsize=(10, 5))
plt.scatter(df['A'], df['B'], alpha=0.7, color='red')
plt.title('Gráfico de dispersión entre A y B')
plt.xlabel('A')
plt.ylabel('B')
plt.grid(True)
plt.show()

# Gráfico de barras para la suma de las columnas A, B, C, y D
df_sum = df[['A', 'B', 'C', 'D']].sum()
plt.figure(figsize=(10, 5))
df_sum.plot(kind='bar', alpha=0.7, color=['blue', 'green', 'red', 'orange'])
plt.title('Suma de las columnas A, B, C y D')
plt.xlabel('Columnas')
plt.ylabel('Suma')
plt.grid(True)
plt.show()
```

```python
#EJEMPLO APLICATIVO 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1 Recopilacion de datos

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(lectura + 'iris.csv')

#2 Limpieza y Preprocesamiento de Datos:

# Mostrar las primeras filas del DataFrame para inspección
print("Primeras filas del DataFrame:")
print(df.head())

# Información básica del DataFrame
print("\nInformación del DataFrame:")
print(df.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(df.describe())

# Verificar valores nulos
print("\nValores nulos en el DataFrame:")
print(df.isnull().sum())

#3 Análisis Exploratorio de Datos (EDA) Y #4 Visualización de Resultados:

# Histograma de cada característica
df.hist(bins=20, figsize=(10, 10))
plt.suptitle("Histogramas de las características del conjunto de datos Iris")
plt.show()

# Boxplot para cada característica
df.plot(kind='box', subplots=True, layout=(2,3), figsize=(12, 8), title="Boxplots de las características del conjunto de datos Iris")
plt.show()

# Pairplot de Seaborn para visualizar relaciones entre características
import seaborn as sns
sns.pairplot(df, hue='Species', markers=["o", "s", "D"])
plt.suptitle("Pairplot del conjunto de datos Iris")
plt.show()

# Análisis de correlación (excluyendo la columna 'Species')
correlation_matrix = df.drop('Species', axis=1).corr()  # Excluimos la columna 'Species'
print("\nMatriz de correlación:")
print(correlation_matrix)

# Mapa de calor de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Mapa de calor de la matriz de correlación")
plt.show()

# Análisis de correlación con scatter plots
sns.scatterplot(data=df, x='Sepal.Length', y='Sepal.Width', hue='Species') # Use 'Sepal.Length' instead of 'sepal_length'
plt.title("Scatter plot de Sepal Length vs Sepal Width")
plt.show()

sns.scatterplot(data=df, x='Petal.Length', y='Petal.Width', hue='Species') # Use 'Petal.Length' instead of 'petal_length'
plt.title("Scatter plot de Petal Length vs Petal Width")
plt.show()

# Análisis de valores promedio por especie
print("\nPromedio de características por especie:")
print(df.groupby('Species').mean()) # Change 'species' to 'Species'

# Visualización de los promedios por especie
df.groupby('Species').mean().plot(kind='bar', figsize=(10, 6)) # Change 'species' to 'Species'
plt.title("Promedio de características por especie")
plt.xlabel("Especie")
plt.ylabel("Promedio")
plt.xticks(rotation=0)
plt.show()
```

```python
#EJEMPLO APLICATIVO 2

#1 Recopilacion de datos

# Cargar el archivo mtcars.txt en un DataFrame
mtcars_df = pd.read_csv(lectura + "mtcars.txt", sep='\t')

#2 Limpieza y Preprocesamiento de Datos:

# Mostrar las primeras filas del DataFrame para inspección
print("Primeras filas del DataFrame:")
print(mtcars_df.head())

# Información básica del DataFrame
print("\nInformación del DataFrame:")
print(mtcars_df.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(mtcars_df.describe())

# Verificar valores nulos
print("\nValores nulos en el DataFrame:")
print(mtcars_df.isnull().sum())

#3 Análisis Exploratorio de Datos (EDA) Y #5 Visualización de Resultados:

# Histograma de cada característica
mtcars_df.hist(bins=20, figsize=(12, 10))
plt.suptitle("Histogramas de las características del conjunto de datos mtcars")
plt.show()

# Boxplot para cada característica
mtcars_df.plot(kind='box', subplots=True, layout=(4,4), figsize=(15, 12), title="Boxplots de las características del conjunto de datos mtcars")
plt.show()

# Pairplot de Seaborn para visualizar relaciones entre características
sns.pairplot(mtcars_df)
plt.suptitle("Pairplot del conjunto de datos mtcars")
plt.show()

# Análisis de correlación (excluyendo la columna 'model')
correlation_matrix = mtcars_df.select_dtypes(include=['float64', 'int64']).corr() # Select only numeric columns for correlation
print("\nMatriz de correlación:")
print(correlation_matrix)

# Mapa de calor de la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Mapa de calor de la matriz de correlación")
plt.show()

# Análisis de correlación con scatter plots
sns.scatterplot(data=mtcars_df, x='mpg', y='hp')
plt.title("Scatter plot de mpg vs hp")
plt.show()

sns.scatterplot(data=mtcars_df, x='wt', y='mpg')
plt.title("Scatter plot de wt vs mpg")
plt.show()

# Análisis de valores promedio por número de cilindros
print("\nPromedio de características por número de cilindros:")
# Exclude the 'model' column when calculating the mean
print(mtcars_df.drop('model', axis=1).groupby('cyl').mean())

# Visualización de los promedios por número de cilindros
mtcars_df.drop('model', axis=1).groupby('cyl').mean().plot(kind='bar', figsize=(12, 8)) # Exclude 'model' here as well
plt.title("Promedio de características por número de cilindros")
plt.xlabel("Número de cilindros")
plt.ylabel("Promedio")
plt.xticks(rotation=0)
plt.show()
```
